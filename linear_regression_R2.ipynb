library import

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn

np.random.seed(42)
np.set_printoptions(precision=6, suppress=True, edgeitems=30, linewidth=120, 
                    formatter=dict(float=lambda x: f'{x:.2f}'))
sns.set(font_scale=1.3)
```

Generate data

```python
from sklearn.datasets import make_regression

data, target = make_regression(n_samples=100, n_features=1, n_targets=1, 
                               noise=30.0, random_state=42)

print(f'data shape: {data.shape}')
print(f'target shape: {target.shape}')
```

```python
data[:5]
```

```python
target[:5]
```

```python
plt.figure(figsize=(8, 6))
plt.title('Regresja liniowa')
plt.xlabel('cecha x')
plt.ylabel('zmienna docelowa')
plt.scatter(data, target, label='cecha x')
plt.legend()
plt.plot()
```

Regresja liniowa przy użyciu scikit-learn

```python
from sklearn.linear_model import LinearRegression

regression = LinearRegression()
regression.fit(data, target)

print(regression.intercept_)
print(regression.coef_[0])
```

```python
regression.score(data, target)
```

```python
y_pred = regression.predict(data)
y_pred
```

Wizualizacja gradiczna modelu

```python
plt.figure(figsize=(8, 6))
plt.title('Regresja liniowa')
plt.xlabel('cecha x')
plt.ylabel('zmienna docelowa')
plt.scatter(data, target, label='cecha x')
plt.plot(data, y_pred, color='red', label='model')
plt.legend()
plt.show()
```

Regresja z podziałem na zbiory treningowy i testowy

```python
data, target = make_regression(n_samples=1000, n_features=1, n_targets=1, noise=15.0, random_state=42)

print(f'data shape: {data.shape}')
print(f'target shape: {target.shape}')
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.25)

print(f'X_train shape: {X_train.shape}')
print(f'y_train shape: {y_train.shape}')
print(f'X_test shape: {X_test.shape}')
print(f'y_test shape: {y_test.shape}')
```

```python
plt.figure(figsize=(8, 6))
plt.title('Regresja liniowa - zbiory')
plt.xlabel('cecha x')
plt.ylabel('zmienna docelowa')
plt.scatter(X_train, y_train, label='zbiór treningowy', color='gray', alpha=0.5)
plt.scatter(X_test, y_test, label='zbiór testowy', color='gold', alpha=0.5)
plt.legend()
plt.show()
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
regressor.score(X_train, y_train)
```

```python
regressor.score(X_test, y_test)
```

Regresja liniowa - zbiór treningowy - wizualizacja

```python
plt.figure(figsize=(8, 6))
plt.title('Regresja liniowa')
plt.xlabel('cecha x')
plt.ylabel('zmienna docelowa')
plt.scatter(X_train, y_train, label='cecha x', color='gray', alpha=0.5)
plt.plot(X_train, regressor.intercept_ + (regressor.coef_ * X_train), color='red', label='model')
plt.legend()
plt.show()
```

Regresja liniowa - zbiór testowy - wizualizacja

```python
plt.figure(figsize=(8, 6))
plt.title('Regresja liniowa')
plt.xlabel('cecha x')
plt.ylabel('zmienna docelowa')
plt.scatter(X_test, y_test, label='cecha x', color='gray', alpha=0.5)
plt.plot(X_test, regressor.intercept_ + (regressor.coef_ * X_test), color='red', label='model')
plt.legend()
plt.show()
```

Predykcja na podstawie modelu

```python
y_pred = regressor.predict(X_test)

predictions = pd.DataFrame(data={'y_true': y_test, 'y_pred': y_pred})
predictions.head()
```

```python
predictions['error'] = predictions['y_true'] - predictions['y_pred']
predictions['error'].head()
```
