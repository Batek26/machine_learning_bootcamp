Import library

```python
import sklearn
import pandas as pd
import numpy as np
```

Generate data

```python
data = {
    'size': ['XL', 'L', 'M', 'L', 'M'],
    'color': ['red', 'green', 'blue', 'green', 'red'],
    'gender': ['female', 'male', 'male', 'female', 'female'],
    'price': [199.0, 89.0, 99.0, 129.0, 79.0],
    'weight': [500, 450, 300, 380, 410],
    'bought': ['yes', 'no', 'yes', 'no', 'yes']
}

df_raw = pd.DataFrame(data=data)
df_raw
```

Copy data

```python
df = df_raw.copy()
df.info()
```

Zmiana typu danych

```python
for col in ['size', 'color', 'gender', 'bought']:
  df[col] = df[col].astype('category')

df['weight'] = df['weight'].astype('float')

df.info()
```

```python
df.describe()
```

```python
df.describe().T
```

```python
df.describe(include=['category']).T
```

LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
le.fit(df['bought'])
le.transform(df['bought'])
```

```python
le.fit_transform(df['bought'])
```

```python
le.classes_
```

```python
df['bought'] = le.fit_transform(df['bought'])
df
```

```python
le.inverse_transform(df['bought'])
```

OneHotEncoder

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse=False)
encoder.fit(df[['size']])
```

```python
encoder.transform(df[['size']])
```

```python
encoder.categories_
```

```python
encoder = OneHotEncoder(drop='first', sparse=False)
encoder.fit_transform(df[['size']])
```

Pandas get_dummies()

```python
pd.get_dummies(data=df)
```

```python
pd.get_dummies(data=df, drop_first=True)
```

```python
pd.get_dummies(data=df, drop_first=True, prefix='new', prefix_sep='-')
```

```python
pd.get_dummies(data=df, drop_first=True, columns=['size'])
```

Standaryzacja

```python
print(f"{df['price']}\n")
print(f"rednia: {df['price'].mean()}\n")
print(f"Odchylenie standardowe: {df['price'].std():.4f}\n")
```

```python
(df['price'] - df['price'].mean()) / df['price'].std()
```

```python
def standarize(x):
  return ((x - x.mean()) / x.std())
```

```python
standarize(df['price'])
```

```python
from sklearn.preprocessing import scale

scale(df['price'])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit_transform(df[['price']])
```

Final data processing

```python
le = LabelEncoder()
df['bought'] = le.fit_transform(df['bought'])

scaler_price = StandardScaler()
df[['price', 'weight']] = scaler_price.fit_transform(df[['price', 'weight']])

df = pd.get_dummies(data=df, drop_first=True)

df
```
