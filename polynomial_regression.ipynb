library import

```python
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

np.random.seed(42)
np.set_printoptions(precision=6, suppress=True)
sns.set(font_scale=1.3)
```

Generate data

```python
X = np.arange(-10, 10, 0.5)
noise = 80 * np.random.randn(40)
y = -X**3 + 10*X**2 - 2*X + 3 + noise
X = X.reshape(40, 1)

plt.figure(figsize=(8, 6))
plt.title('Regresja wielomianowa')
plt.xlabel('cecha x')
plt.ylabel('zmienna docelowa')
plt.scatter(X, y, label='cecha x')
plt.legend()
plt.show()
```

Model liniowy

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(X, y)
y_pred_line = regressor.predict(X)

plt.figure(figsize=(8, 6))
plt.title('Regresja wielomianowa')
plt.xlabel('cecha x')
plt.ylabel('zmienna docelowa')
plt.scatter(X, y, label='cecha x')
plt.plot(X, y_pred_line, color='red', label='regresja liniowa')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import r2_score

r2_score(y, y_pred_line)
```

Ekstrakcja cech wielomianowych

```python
df = pd.DataFrame(data={'X': X.ravel()})
df.head(10)
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2)

X_poly = poly.fit_transform(X)
X_poly
```

```python
X_poly.shape
```

```python
df = pd.DataFrame(X_poly)
df.columns = ['1', 'x', 'x^2']
df.head()
```

Dopasowanie modelu

```python
regressor_poly = LinearRegression()
regressor_poly.fit(X_poly, y)
y_pred_2 = regressor_poly.predict(X_poly)

plt.figure(figsize=(8, 6))
plt.title('Regresja wielomianowa')
plt.xlabel('cecha x')
plt.ylabel('zmienna docelowa')
plt.scatter(X, y, label='cecha x')
plt.plot(X, y_pred_line, color='red', label='regresja liniowa')
plt.plot(X, y_pred_2, color='green', label='regresja wielomianowa - stopien 2')
plt.legend()
plt.show()
```

```python
r2_score(y, y_pred_2)
```

Ekstrakcja cech wielomianowych - stopien 3

```python
poly = PolynomialFeatures(degree=3)

X_poly_3 = poly.fit_transform(X)
X_poly_3
```

Dopasowanie modelu

```python
df = pd.DataFrame(X_poly_3)
df.columns = ['1', 'x', 'x^2', 'x^3']
df.head()
```

```python
regressor_poly = LinearRegression()
regressor_poly.fit(X_poly_3, y)
y_pred_3 = regressor_poly.predict(X_poly_3)

plt.figure(figsize=(8, 6))
plt.title('Regresja wielomianowa')
plt.xlabel('cecha x')
plt.ylabel('zmienna docelowa')
plt.scatter(X, y, label='cecha x')
plt.plot(X, y_pred_line, color='red', label='regresja liniowa')
plt.plot(X, y_pred_2, color='green', label='regresja wielomianowa - stopien 2')
plt.plot(X, y_pred_3, color='gold', label='regresja wielomianowa - stopien 3')
plt.legend()
plt.show()
```

```python
r2_score(y, y_pred_3)
```

Porwnanie metryk

```python
from sklearn.metrics import mean_absolute_error as mae
from sklearn.metrics import mean_squared_error as mse
```

```python
results = pd.DataFrame(data={
    'name': ['regresja liniowa', 'regresja wielomianowa st.2', 'regresja wielomianowa st.3'],
    'r2_score': [r2_score(y, y_pred_line), r2_score(y, y_pred_2), r2_score(y, y_pred_3)],
    'mae': [mae(y, y_pred_line), mae(y, y_pred_2), mae(y, y_pred_3)],
    'mse': [mse(y, y_pred_line), mse(y, y_pred_2), mse(y, y_pred_3)],
    'rmse': [np.sqrt(mse(y, y_pred_line)), np.sqrt(mse(y, y_pred_2)), np.sqrt(mse(y, y_pred_3))]                           
    })

results
```

```python
px.bar(results, x='name', y='r2_score', width=700, title='Regresja wielomianowa - R2_score')
```

```python
px.bar(results, x='name', y='mae', width=700, title='Regresja wielomianowa - MAE')
```

```python
px.bar(results, x='name', y='mse', width=700, title='Regresja wielomianowa - MSE')
```

```python
px.bar(results, x='name', y='rmse', width=700, title='Regresja wielomianowa - RMSE')
```
